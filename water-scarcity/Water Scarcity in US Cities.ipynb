```python
# Load the combined GHCN data file
ghcn_path = "/Users/justinpizzoferrato/Desktop/DCS680/Water/Data/ghcn_combined_1985_2025.csv"
ghcn_df = pd.read_csv(ghcn_path)

# Check structure and station names
print("Columns:", ghcn_df.columns)
print("Unique stations:", ghcn_df["NAME"].unique())
ghcn_df.head()
```

```python
# Filter for Fresno station only
fresno_ghcn = ghcn_df[ghcn_df["NAME"].str.contains("FRESNO", case=False)].copy()

# Convert date column to datetime
fresno_ghcn["DATE"] = pd.to_datetime(fresno_ghcn["DATE"])

# Extract year
fresno_ghcn["year"] = fresno_ghcn["DATE"].dt.year

# Convert PRCP from tenths of mm to mm (1 = 0.1 mm)
fresno_ghcn["PRCP_mm"] = pd.to_numeric(fresno_ghcn["PRCP"], errors="coerce") * 0.1

# Group by year and calculate total annual precipitation
fresno_annual_prcp = (
    fresno_ghcn.groupby("year")["PRCP_mm"]
    .sum()
    .reset_index()
    .rename(columns={"PRCP_mm": "precip_mm"})
)

# Add city label
fresno_annual_prcp["City"] = "Fresno"

# Preview result
print(fresno_annual_prcp.shape)
fresno_annual_prcp.tail()
```

```python
# Filter for Houston station only
houston_ghcn = ghcn_df[ghcn_df["NAME"].str.contains("HOUSTON", case=False)].copy()

# Convert date column
houston_ghcn["DATE"] = pd.to_datetime(houston_ghcn["DATE"])
houston_ghcn["year"] = houston_ghcn["DATE"].dt.year

# Convert PRCP from tenths of mm to mm
houston_ghcn["PRCP_mm"] = pd.to_numeric(houston_ghcn["PRCP"], errors="coerce") * 0.1

# Group by year and calculate total annual precipitation
houston_annual_prcp = (
    houston_ghcn.groupby("year")["PRCP_mm"]
    .sum()
    .reset_index()
    .rename(columns={"PRCP_mm": "precip_mm"})
)

# Add city label
houston_annual_prcp["City"] = "Houston"

# Preview result
print(houston_annual_prcp.shape)
houston_annual_prcp.tail()
```

```python
# Filter for Phoenix station only
phoenix_ghcn = ghcn_df[ghcn_df["NAME"].str.contains("PHOENIX", case=False)].copy()

# Convert date column
phoenix_ghcn["DATE"] = pd.to_datetime(phoenix_ghcn["DATE"])
phoenix_ghcn["year"] = phoenix_ghcn["DATE"].dt.year

# Convert PRCP from tenths of mm to mm
phoenix_ghcn["PRCP_mm"] = pd.to_numeric(phoenix_ghcn["PRCP"], errors="coerce") * 0.1

# Group by year and calculate total annual precipitation
phoenix_annual_prcp = (
    phoenix_ghcn.groupby("year")["PRCP_mm"]
    .sum()
    .reset_index()
    .rename(columns={"PRCP_mm": "precip_mm"})
)

# Add city label
phoenix_annual_prcp["City"] = "Phoenix"

# Preview result
print(phoenix_annual_prcp.shape)
phoenix_annual_prcp.tail()
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Filter latest year from merged_df
latest_year = merged_df["year"].max()
latest_data = merged_df[merged_df["year"] == latest_year][["City", "PerCapita_Adjusted"]]

# Merge with ACS summary
profile_df = pd.merge(acs_summary_df, latest_data, on="City", how="left")

# Clean values
profile_df["Median_Income"] = profile_df["Median_Income"].str.replace(",", "").astype(float)
profile_df["Unemployment_Rate"] = profile_df["Unemployment_Rate"].str.replace("%", "").astype(float)
profile_df["Poverty_Rate"] = profile_df["Poverty_Rate"].str.replace("%", "").astype(float)
profile_df["Housing_Cost_Burden"] = profile_df["Housing_Cost_Burden"].str.replace("%", "").astype(float)
```

```python
# Metrics to normalize
metrics = ["Median_Income", "Unemployment_Rate", "Poverty_Rate", "Housing_Cost_Burden", "PerCapita_Adjusted"]

# Normalize
scaler = MinMaxScaler()
normalized_values = scaler.fit_transform(profile_df[metrics])
normalized_df = pd.DataFrame(normalized_values, columns=metrics)
normalized_df["City"] = profile_df["City"]

# Reorder columns
radar_df = normalized_df.set_index("City")

# Radar chart setup
labels = radar_df.columns
num_vars = len(labels)

# Compute angle for each axis
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # Complete the loop

# Initialize plot
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# Plot for each city
for i, row in radar_df.iterrows():
    values = row.tolist()
    values += values[:1]  # Close the circle
    ax.plot(angles, values, label=i)
    ax.fill(angles, values, alpha=0.1)

# Axis labels
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels, fontsize=10)
ax.set_yticklabels([])
ax.set_title("City Profiles: Normalized Risk & Water Burden Factors", size=13, pad=20)
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))

plt.tight_layout()
plt.show()
```

```python
# Estimate water withdrawal trends under future years

# Assume `merged_df` contains historical withdrawals and year
trend_df = merged_df.copy()
trend_df = trend_df.dropna(subset=["Withdrawals_Mgal_per_day"])

# Forecast range
future_years = np.arange(2025, 2051)

# Fit linear trend model by city
from sklearn.linear_model import LinearRegression

future_preds = []
for city in trend_df['City'].unique():
    city_data = trend_df[trend_df['City'] == city]
    X = city_data[['year']]
    y = city_data['Withdrawals_Mgal_per_day']
    
    model = LinearRegression().fit(X, y)
    
    preds = model.predict(future_years.reshape(-1, 1))
    future_city_df = pd.DataFrame({
        "City": city,
        "year": future_years,
        "Projected_Withdrawals": preds
    })
    future_preds.append(future_city_df)

future_water_df = pd.concat(future_preds)

# Preview
future_water_df.head()
```

```python
# Add climate stress multiplier (1% decline per year)
base_year = 2025
future_water_df["Climate_Adjustment_Factor"] = 1 - 0.01 * (future_water_df["year"] - base_year)

# Apply to withdrawals
future_water_df["Climate_Adj_Withdrawals"] = (
    future_water_df["Projected_Withdrawals"] * future_water_df["Climate_Adjustment_Factor"]
)

# Preview climate-adjusted scenario
future_water_df.head()
```

```python
# Filter to future years (if not already)
plot_df = future_water_df.copy()

# Plot projections
plt.figure(figsize=(10, 6))
sns.lineplot(data=plot_df, x="year", y="Projected_Withdrawals", hue="City", style="City", markers=True, dashes=False)
sns.lineplot(data=plot_df, x="year", y="Climate_Adj_Withdrawals", hue="City", style="City", markers=True, dashes=True, legend=False)

plt.title("Projected vs Climate-Adjusted Water Withdrawals by City")
plt.ylabel("Withdrawals (Mgal/day)")
plt.xlabel("Year")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Set a reference point for "safe" water use â€” could be baseline 2025
baseline_df = future_water_df[future_water_df["year"] == 2025]
thresholds = baseline_df.set_index("City")["Climate_Adj_Withdrawals"].to_dict()

# Identify when usage exceeds the 2025 baseline
def find_threshold_breach(city_df, threshold):
    over_limit = city_df[city_df["Climate_Adj_Withdrawals"] < threshold]
    return over_limit["year"].min() if not over_limit.empty else None

# Apply
breach_years = {}
for city in future_water_df["City"].unique():
    city_df = future_water_df[future_water_df["City"] == city]
    breach_years[city] = find_threshold_breach(city_df, thresholds[city])

# Display
breach_years
```

```python
# Create a copy for mitigation scenario
mitigation_df = future_water_df.copy()

# Apply 10% reduction starting in 2026
mitigation_df["Mitigated_Withdrawals"] = mitigation_df.apply(
    lambda row: row["Climate_Adj_Withdrawals"] * 0.90 if row["year"] >= 2026 else row["Climate_Adj_Withdrawals"],
    axis=1
)

# Preview result
mitigation_df.head()
```

```python
# Plot: Climate-Adjusted vs Mitigated Withdrawals
plt.figure(figsize=(10, 6))
sns.lineplot(
    data=mitigation_df,
    x="year",
    y="Climate_Adj_Withdrawals",
    hue="City",
    style="City",
    dashes=True,
    linewidth=2.5
)
sns.lineplot(
    data=mitigation_df,
    x="year",
    y="Mitigated_Withdrawals",
    hue="City",
    style="City",
    dashes=False,
    linewidth=2.5
)

plt.title("Mitigation Scenario: Projected Withdrawals with Climate Stress vs. 10% Conservation")
plt.ylabel("Withdrawals (Mgal/day)")
plt.xlabel("Year")
plt.legend(title="City & Scenario")
plt.tight_layout()
plt.show()
```

```python
# Calculate annual savings (baseline - mitigated)
mitigation_df["Water_Saved"] = (
    mitigation_df["Climate_Adj_Withdrawals"] - mitigation_df["Mitigated_Withdrawals"]
)

# Aggregate total savings by city
savings_summary = (
    mitigation_df.groupby("City")
    .agg(
        Total_Saved_Mgal_day=("Water_Saved", "sum"),
        Avg_Annual_Saved_Mgal_day=("Water_Saved", "mean"),
        Max_Annual_Saved_Mgal_day=("Water_Saved", "max"),
        Min_Annual_Saved_Mgal_day=("Water_Saved", "min")
    )
    .round(2)
    .reset_index()
)

# Display savings summary
savings_summary
```

```python
# Assign price per million gallons
price_per_mgal = 1500  # $/Mgal

# Add value column to summary
savings_summary["Estimated_Value_USD"] = (
    savings_summary["Total_Saved_Mgal_day"] * price_per_mgal
).round(2)

# Preview value-added summary
savings_summary
```

```python
# Calculate difference if usage exceeds the 2025 safe baseline (unmitigated)
shortage_df = future_water_df.copy()

# Safe baseline by city (from 2025)
safe_thresholds = baseline_df.set_index("City")["Climate_Adj_Withdrawals"].to_dict()

# Flag and quantify overuse by year
shortage_df["Overuse_Mgal_day"] = shortage_df.apply(
    lambda row: row["Climate_Adj_Withdrawals"] - safe_thresholds[row["City"]]
    if row["Climate_Adj_Withdrawals"] > safe_thresholds[row["City"]] else 0,
    axis=1
)

# Summarize total and average overuse by city
shortage_summary = shortage_df.groupby("City")["Overuse_Mgal_day"].agg(
    Total_Overuse_Mgal_day="sum",
    Avg_Annual_Overuse_Mgal_day="mean",
    Years_Over_Threshold=lambda x: (x > 0).sum()
).reset_index()

shortage_summary
```

```python
# Define cost of alternative water sources (e.g., desalination, reuse)
replacement_cost_per_mgal = 3000  # higher than standard supply cost

# Estimate value of mitigation in Phoenix
shortage_summary["Overuse_Value_USD"] = (
    shortage_summary["Total_Overuse_Mgal_day"] * replacement_cost_per_mgal
).round(2)

shortage_summary
```

```python
# Combine key variables
final_risk_df = profile_df[["City", "Equity_Risk_Index", "PerCapita_Adjusted"]].copy()

# Merge with shortage info
final_risk_df = final_risk_df.merge(shortage_summary[["City", "Years_Over_Threshold"]], on="City", how="left")
final_risk_df["Years_Over_Threshold"] = final_risk_df["Years_Over_Threshold"].fillna(0)

# Score each city on a 0â€“3 scale for each column
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

for col in ["Equity_Risk_Index", "PerCapita_Adjusted", "Years_Over_Threshold"]:
    final_risk_df[f"{col}_Scaled"] = scaler.fit_transform(final_risk_df[[col]])

# Composite score (equal weights)
final_risk_df["Composite_Risk_Score"] = (
    final_risk_df[[f"{col}_Scaled" for col in ["Equity_Risk_Index", "PerCapita_Adjusted", "Years_Over_Threshold"]]]
    .sum(axis=1)
    .round(2)
)

# Sort by highest risk
final_risk_df.sort_values("Composite_Risk_Score", ascending=False)
```
